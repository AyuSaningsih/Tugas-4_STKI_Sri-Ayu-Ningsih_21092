```python
# Inisialisasi dictionary untuk menyimpan daftar belanja
daftar_belanja = {}

# Fungsi untuk menambahkan barang ke daftar belanja
def tambah_barang(daftar, barang, jumlah):
    if barang in daftar:
        daftar[barang] += jumlah
    else:
        daftar[barang] = jumlah

# Fungsi untuk menghapus barang dari daftar belanja
def hapus_barang(daftar, barang):
    if barang in daftar:
        del daftar[barang]
    else:
        print("Barang tidak ditemukan dalam daftar belanja.")

# Fungsi untuk menampilkan daftar belanja
def tampilkan_daftar(daftar):
    print("Daftar Belanja:")
    for barang, jumlah in daftar.items():
        print(f"{barang}: {jumlah}")

# Menu utama
while True:
    print("\nMenu:")
    print("1. Tambah Barang")
    print("2. Hapus Barang")
    print("3. Tampilkan Daftar Belanja")
    print("4. Keluar")
    pilihan = input("Pilih menu: ")

    if pilihan == '1':
        barang = input("Masukkan nama barang: ")
        jumlah = int(input("Masukkan jumlah: "))
        tambah_barang(daftar_belanja, barang, jumlah)
        print(f"{barang} ditambahkan ke daftar belanja.")
    elif pilihan == '2':
        barang = input("Masukkan nama barang yang ingin dihapus: ")
        hapus_barang(daftar_belanja, barang)
    elif pilihan == '3':
        tampilkan_daftar(daftar_belanja)
    elif pilihan == '4':
        print("Terima kasih! Program selesai.")
        break
    else:
        print("Pilihan tidak valid. Silakan pilih 1-4.")
```

```python
class Simpul:
    def __init__(self, nama):
        self.nama = nama
        self.anak = []

    def tambah_anak(self, anak):
        self.anak.append(anak)

    def tampilkan(self, level=0):
        indentasi = ' ' * level * 2
        print(f"{indentasi}{self.nama}")
        for anak in self.anak:
            anak.tampilkan(level + 1)

# Membangun pohon genealogi
budi = Simpul("Budi")
ani = Simpul("Ani")
citra = Simpul("Citra")
deni = Simpul("Deni")
efi = Simpul("Efi")
fera = Simpul("Fera")

budi.tambah_anak(ani)
budi.tambah_anak(deni)
ani.tambah_anak(citra)
ani.tambah_anak(efi)
deni.tambah_anak(fera)

# Menampilkan pohon genealogi
budi.tampilkan()
```
